##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
# Drop missing values from ConvertedCompYearly
comp_data = df['ConvertedCompYearly'].dropna()

# Set figure size
plt.figure(figsize=(10, 6))

# Plot histogram with KDE
sns.histplot(comp_data, bins=50, kde=True, color="skyblue")

# Add labels and title
plt.title("Distribution of Yearly Compensation", fontsize=16)
plt.xlabel("Yearly Compensation (USD)", fontsize=12)
plt.ylabel("Number of Respondents", fontsize=12)

# Show the plot
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter for full-time employees and drop missing compensation values
full_time_comp = df[df['Employment'] == "Employed, full-time"]['ConvertedCompYearly'].dropna()

# Calculate the median compensation
median_full_time = full_time_comp.median()

print("Median yearly compensation for full-time employees:", median_full_time)
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
# Consider only rows with non-missing Country and ConvertedCompYearly
df_country_comp = df[['Country', 'ConvertedCompYearly']].dropna()

# Optional: Focus on top N countries by respondent count for clarity
top_countries = df_country_comp['Country'].value_counts().nlargest(10).index
df_top_countries = df_country_comp[df_country_comp['Country'].isin(top_countries)]

# Plot boxplot
plt.figure(figsize=(12,6))
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_top_countries)
plt.xticks(rotation=45)
plt.ylim(0, 300000)  # Optional: limit y-axis to remove extreme outliers for readability
plt.title('Distribution of Yearly Compensation by Country (Top 10 Countries)')
plt.ylabel('Yearly Compensation (USD)')
plt.xlabel('Country')
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
# Consider only rows with non-missing ConvertedCompYearly
df_comp = df[['ConvertedCompYearly']].dropna()

# Calcolare il primo (Q1) e terzo (Q3) quartile
Q1 = df_comp['ConvertedCompYearly'].quantile(0.25)
Q3 = df_comp['ConvertedCompYearly'].quantile(0.75)

# Calcolare l'IQR
IQR = Q3 - Q1

# Definire i limiti inferiori e superiori
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Creare un nuovo DataFrame senza outlier
df_no_outliers = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Verificare le dimensioni del nuovo DataFrame
print(f"Original dataset size: {df.shape}")
print(f"Refined dataset size (outliers removed): {df_no_outliers.shape}")
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
# Selezionare le colonne di interesse
corr_data = df_no_outliers[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']]

# Calcolare la matrice di correlazione
corr_matrix = corr_data.corr()

# Stampare la matrice di correlazione
print(corr_matrix)

# Visualizzare la heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Compensation, Work Experience, Job Satisfaction')
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here
# Impostare la dimensione della figura
plt.figure(figsize=(12,5))

# Scatter plot: ConvertedCompYearly vs WorkExp
plt.subplot(1,2,1)
sns.scatterplot(data=df_no_outliers, x='WorkExp', y='ConvertedCompYearly')
plt.title('ConvertedCompYearly vs Work Experience')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Converted Compensation Yearly')

# Scatter plot: ConvertedCompYearly vs JobSatPoints_1
plt.subplot(1,2,2)
sns.scatterplot(data=df_no_outliers, x='JobSatPoints_1', y='ConvertedCompYearly')
plt.title('ConvertedCompYearly vs Job Satisfaction')
plt.xlabel('Job Satisfaction Points')
plt.ylabel('Converted Compensation Yearly')

plt.tight_layout()
plt.show()
```
